# Tipos de Datos en Python

## 1. Enteros (int)
Representan números enteros.

### Operaciones comunes:
- `+`, `-`, `*`, `/`, `%`, `//`, `**`

### Funciones útiles:
- `abs(x)`: Valor absoluto
- `pow(x, y)`: x elevado a la potencia y
- `int(x)`: Convierte a entero

---

## 2. Flotantes (float)
Representan números decimales.

### Operaciones comunes:
- `+`, `-`, `*`, `/`, `**`

### Funciones útiles:
- `round(x, n)`: Redondea x con n decimales
- `float(x)`: Convierte a flotante

---

## 3. Cadenas (str)
Secuencias de caracteres.

### Operaciones comunes:
- `+` (concatenar)
- `*` (repetir)

### Métodos útiles:
- `len(s)`: Longitud de la cadena
- `str(x)`: Convierte a cadena
- `s.lower()`: Convierte a minúsculas
- `s.upper()`: Convierte a mayúsculas
- `s.capitalize()`: Capitaliza la primera letra
- `s.strip()`: Elimina espacios en blanco
- `s.replace(old, new)`: Reemplaza texto
- `s.split()`: Divide la cadena en palabras
- `s.join(iterable)`: Une un iterable en una cadena

---

## 4. Listas (list)
Secuencias ordenadas y mutables.

### Métodos útiles:
- `len(lista)`: Longitud de la lista
- `lista.append(x)`: Añade un elemento al final
- `lista.insert(i, x)`: Inserta un elemento en la posición i
- `lista.pop(i)`: Elimina el elemento en la posición i
- `lista.remove(x)`: Elimina la primera aparición de x
- `lista.sort()`: Ordena la lista
- `lista.reverse()`: Invierte la lista

---

## 5. Tuplas (tuple)
Secuencias ordenadas e inmutables.

### Métodos útiles:
- `len(tupla)`: Longitud de la tupla
- `tupla.index(x)`: Índice de la primera aparición de x
- `tupla.count(x)`: Cuenta las veces que aparece x

---

## 6. Diccionarios (dict)
Colecciones de pares clave-valor.

### Métodos útiles:
- `len(diccionario)`: Longitud del diccionario
- `diccionario.get(clave)`: Obtiene el valor de la clave
- `diccionario.keys()`: Devuelve las claves
- `diccionario.values()`: Devuelve los valores
- `diccionario.items()`: Devuelve pares clave-valor
- `diccionario.pop(clave)`: Elimina una clave
- `diccionario.update(otro_diccionario)`: Actualiza el diccionario

---

## 7. Sets (set)
Colecciones desordenadas de elementos únicos.

### Operaciones comunes:
- `|` (unión)
- `&` (intersección)
- `-` (diferencia)
- `^` (diferencia simétrica)

### Métodos útiles:
- `len(set)`: Longitud del set
- `set.add(x)`: Añade un elemento
- `set.remove(x)`: Elimina un elemento (error si no existe)
- `set.discard(x)`: Elimina un elemento (sin error si no existe)
- `set.pop()`: Elimina y devuelve un elemento aleatorio
- `set.clear()`: Vacía el set

---

## 8. Booleanos (bool)
Representan `True` o `False`.

### Operadores lógicos:
- `and`, `or`, `not`

### Comparaciones:
- `==`, `!=`, `<`, `>`, `<=`, `>=`

---

## 9. None
Representa la ausencia de valor.

### Usos comunes:
- Inicializar variables o como valor por defecto en funciones
